```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
from sklearn.datasets import fetch_olivetti_faces
faces = fetch_olivetti_faces()
print(faces.DESCR)
```

```python
len(faces['data'])
```

```python
x = faces['data']
y = faces['target']
```

```python
n, k = 4, 10
fig, axes = plt.subplots(n, k, sharex=True, sharey=True)
fig.set_figheight(15)
fig.set_figwidth(15)
for i in range(n*k):
    axes[i//k, i%k].imshow(faces['images'][i], cmap=plt.cm.bone)
```

```python
plt.imshow(faces['images'][0], cmap=plt.cm.bone)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2,
                                                   random_state=True)
```

```python
cls = svm.SVC()
cls.fit(x_train, y_train)
y_pred = cls.predict(x_test)

print('  :')
print(metrics.accuracy_score(y_test, y_pred))
print(' :')
print(metrics.precision_score(y_test, y_pred, average='macro'))
print(' ')
print(metrics.recall_score(y_test, y_pred, average='macro'))
```
